```python
import pandas as pd
```

```python
df = pd.read_csv("repositories/PDSgroupR_ExamTemplate/data/Annotations.csv", sep=";")
```

```python
from sklearn.metrics import cohen_kappa_score
```

```python
df.columns
```

```python
print(f"Pen; Mihnea and Alex: \t {cohen_kappa_score(df['Mihnea/pen'], df['Alex/pen'])}")
print(f"Pen; Mihnea and Em: \t {cohen_kappa_score(df['Mihnea/pen'], df['Em/pen'])}")
print(f"Pen; Mihnea and Nusrat:  {cohen_kappa_score(df['Mihnea/pen'], df['Nusrat/pen'])}")
print(f"Pen; Mihnea and Sidra: \t {cohen_kappa_score(df['Mihnea/pen'], df['Sidra/pen'])}")
print(f"Pen; Alex and Em: \t {cohen_kappa_score(df['Alex/pen'], df['Em/pen'])}")
print(f"Pen; Alex and Nusrat: \t {cohen_kappa_score(df['Alex/pen'], df['Nusrat/pen'])}")
print(f"Pen; Alex and Sidra: \t {cohen_kappa_score(df['Alex/pen'], df['Sidra/pen'])}")
print(f"Pen; Em and Nusrat: \t {cohen_kappa_score(df['Em/pen'], df['Nusrat/pen'])}")
print(f"Pen; Em and Sidra: \t {cohen_kappa_score(df['Em/pen'], df['Sidra/pen'])}")
print(f"Pen; Nusrat and Sidra: \t {cohen_kappa_score(df['Nusrat/pen'], df['Sidra/pen'])}")
```

```python
print(f"Hair; Mihnea and Alex: \t {cohen_kappa_score(df['Mihnea/hair'], df['Alex/hair'], weights="linear")}")
print(f"Hair; Mihnea and Em: \t {cohen_kappa_score(df['Mihnea/hair'], df['Em/hair'], weights="linear")}")
print(f"Hair; Mihnea and Nusrat: {cohen_kappa_score(df['Mihnea/hair'], df['Nusrat/hair'], weights="linear")}")
print(f"Hair; Mihnea and Sidra:  {cohen_kappa_score(df['Mihnea/hair'], df['Sidra/hair'], weights="linear")}")
print(f"Hair; Alex and Em: \t {cohen_kappa_score(df['Alex/hair'], df['Em/hair'], weights="linear")}")
print(f"Hair; Alex and Nusrat: \t {cohen_kappa_score(df['Alex/hair'], df['Nusrat/hair'], weights="linear")}")
print(f"Hair; Alex and Sidra: \t {cohen_kappa_score(df['Alex/hair'], df['Sidra/hair'], weights="linear")}")
print(f"Hair; Em and Nusrat: \t {cohen_kappa_score(df['Em/hair'], df['Nusrat/hair'], weights="linear")}")
print(f"Hair; Em and Sidra: \t {cohen_kappa_score(df['Em/hair'], df['Sidra/hair'], weights="linear")}")
print(f"Hair; Nusrat and Sidra:  {cohen_kappa_score(df['Nusrat/hair'], df['Sidra/hair'], weights="linear")}")
```
